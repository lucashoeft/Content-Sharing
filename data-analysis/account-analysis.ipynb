```python
import pandas as pd
from IPython.display import display
import seaborn as sns
```

```python
# Includes all political members of the parliament
mdb_list = pd.read_csv('../data/processed/mdb_list/mdb_list.csv', sep=";", na_values="")

# Include all found twitter profiles
user_list = pd.read_csv('../data/processed/user_list/user_list.csv', sep=";", na_values="", dtype={'twitter_id': str})

# Total numbers of politicians
print(round(user_list.shape[0] / mdb_list.shape[0] * 100,2), "% of MdB use Twitter")

# Break down by party
print(round(user_list[user_list['fraktion'] == 'Fraktion DIE LINKE.'].shape[0] / mdb_list[mdb_list['fraktion'] == 'Fraktion DIE LINKE.'].shape[0] * 100, 2), "% of Linke use Twitter")
print(round(user_list[user_list['fraktion'] == 'Fraktion der Christlich Demokratischen Union/Christlich - Sozialen Union'].shape[0] / mdb_list[mdb_list['fraktion'] == 'Fraktion der Christlich Demokratischen Union/Christlich - Sozialen Union'].shape[0] * 100, 2), "% of CDU/CSU use Twitter")
print(round(user_list[user_list['fraktion'] == 'Fraktion BÜNDNIS 90/DIE GRÜNEN'].shape[0] / mdb_list[mdb_list['fraktion'] == 'Fraktion BÜNDNIS 90/DIE GRÜNEN'].shape[0] * 100, 2), "% of Die Grünen use Twitter")
print(round(user_list[user_list['fraktion'] == 'Fraktion der Sozialdemokratischen Partei Deutschlands'].shape[0] / mdb_list[mdb_list['fraktion'] == 'Fraktion der Sozialdemokratischen Partei Deutschlands'].shape[0] * 100, 2), "% of SDP use Twitter")
print(round(user_list[user_list['fraktion'] == 'Fraktion der Freien Demokratischen Partei'].shape[0] / mdb_list[mdb_list['fraktion'] == 'Fraktion der Freien Demokratischen Partei'].shape[0] * 100, 2), "% of FDP use Twitter")
print(round(user_list[user_list['fraktion'] == 'Alternative für Deutschland'].shape[0] / mdb_list[mdb_list['fraktion'] == 'Alternative für Deutschland'].shape[0] * 100, 2), "% of AFD use Twitter")
print(round(user_list[user_list['fraktion'] == 'Fraktionslos'].shape[0] / mdb_list[mdb_list['fraktion'] == 'Fraktionslos'].shape[0] * 100, 2), "% of Fraktionslos use Twitter")
```

```python
ax3 = sns.countplot(x="fraktion", data=mdb_list.sort_values(by="fraktion"))
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=90)
ax3.plot()
```

```python
ax4 = sns.countplot(x="fraktion", data=user_list.sort_values(by="fraktion"))
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=90)
ax4.plot()
```
